# Matplotlib Line Charts

Line graphs are used to track changes over time. When smaller changes exist, line graphs are better to use than bar graphs.

Let's create a simple line chart representing revenue over each quarter in time using:
<br>
*revenue* = [200, 400, 650, 800, 850]
<br><br>
Remember to import the `pyplot` module from Matplotlib.

```python
import matplotlib.pyplot as plt
revenue = [200, 400, 650, 800, 850]
time = range(5)
plt.plot(time, revenue)
plt.show()
```

The syntax here is just like we saw with the scatter plot, except we use **plt.plot(x, y)** to return a line chart representing the data.
<br><br>
Just like we saw with the scatter plot, you can add additional functions to add features to your chart.

```python
revenue = [200, 400, 650, 800, 850]
time = range(5)
plt.plot(time, revenue)
plt.title("This is My Line Chart")
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.legend(["Revenue for Company A"])
plt.show()
```

# Multiple Line Plots On the Same Chart

Line graphs can also be used to compare changes over the same period of time for more than one group.  Let's create a line chart representing student attendance for College A and for College B using the following data:
<br><br>
*attendance_A* = [3000, 3005, 3010, 2900, 2950, 3050]
<br>
*attendance_B* = [3050, 3100, 2980, 2980, 2920, 3010]
<br>
*months* = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

```python
attendance_A = [3000, 3005, 3010, 2900, 2950, 3050] 
attendance_B = [3050, 3100, 2980, 2980, 2920, 3010] 
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
months_numbered = range(len(months))
```

The step above defines the data for my line chart.  Notice the *months_numbered* list.  Since we are charting a line graph, the **data must be numerical**.  The *months* list is not numerical.  We can fix this by applying both the **len( )** and **range( )** functions.  What this does is:
<br>
- First, counts the number of items in the months list.
- Then, generates a new list *months_numbered* of numbers from 0 up to but not including the length of *months*
<br><br>
This new list will now represent the period of time for the attendance counts.  Let's print this new list to the screen to see how it looks:

```python
for month in months_numbered:
    print(month)
```

It is perfectly fine to have your time period begin with 0, however you can modify this to begin with 1 if you like.  Let's continue plotting the line chart of student attendance for both colleges:

```python
plt.plot(months_numbered, attendance_A)
plt.plot(months_numbered, attendance_B)
plt.show()
```

Nice.  To plot multiple groups on a line chart, you simply call the **plt.plot( )** function on each group's data you want to graph.  Now let's add features to our chart.

```python
plt.plot(months_numbered, attendance_A)
plt.plot(months_numbered, attendance_B)
plt.title("Student Attendace from January to June")
plt.xlabel("Month")
plt.ylabel("Attendance Counts")
plt.legend(["College A", "College B"])
plt.show()
```

Notice **plt.legend( )** is passed labels for both lines.  This makes it easy to see which line is for College A and which line is for College B.
