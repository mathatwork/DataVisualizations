# Introduction to Matplotlib

Matplotlib is a Python 2D plotting library which produces publication quality figures.  It has a module named `pyplot` which makes things easy for plotting by providing features to control line styles, font properties, formatting axes etc. 
<br><br>
To use Matplotlib, you must ALWAYS first import the package at the beginning of your Python script. 

```python
import matplotlib.pyplot as plt
```

This import is done slightly different than for Numpy and Pandas.  We are choosing here NOT to import the entire library.  We are only going to import the pyplot module from Matplotlib.

Notice there is no output generated.  This is good because that means the module imported successfully.  Also, since we imported pyplot as *plt*, to use the module we simply call *plt*.

# Matplotlib Scatter Plot

Let's create a simple scatter plot representing someone's spending on food over the past week. First, we will need some data:

```python
money_spent = [10,12,12,10,14,22,24]
days = range(1,8)
for day in days:
    print(day)
```

Notice how we generated my list of days using the **range( )** function.  Now letâ€™s use pyplot to graph.

```python
import matplotlib.pyplot as plt
plt.scatter(days, money_spent)
plt.show()
```

Notice the syntax.
- **plt.scatter(x, y)** takes in an array of *x* values and an array of *y* values and then returns a scatter plot representing the data.
- **plt.show( )** is how we print the plot to the screen.  Without it, the graph will NOT render to the screen.

This is really as basic as you can get, but still pretty nice.  You can add things to your scatter plot such as a plot title, *x* and *y*-axis labels, and a legend by calling additional functions BEFORE calling **plt.show( )**.  This is very important because once you call **plt.show( )** the graph will then render to the screen.

**plt.title( )** adds a title to your scatter plot.

```python
plt.scatter(days, money_spent)
plt.title('This is My Scatter Plot')
plt.show()
```

**plt.xlabel( )** adds a label to your *x*-axis.

```python
plt.scatter(days, money_spent)
plt.title('This is My Scatter Plot')
plt.xlabel('Day of the Week')
plt.show()
```

**plt.ylabel( )** adds a label to your *y*-axis.

```python
plt.scatter(days, money_spent)
plt.title('This is My Scatter Plot')
plt.xlabel('Day of the Week')
plt.ylabel('Amount of Money Spent')
plt.show()
```

**plt.legend( )** adds a legend to your plot with the label passed to it.

```python
plt.scatter(days, money_spent)
plt.title('This is My Scatter Plot')
plt.xlabel('Day of the Week')
plt.ylabel('Amount of Money Spent')
plt.legend(['Money Spent'])
plt.show()
```
