# Matplotlib Pie Charts

Pie charts are best to use when you are trying to compare parts of a whole. They do not show changes over time.

Let's create a simple pie chart representing payment method frequency for each payment type using:
<br>
*method_type* = ["Debit Card", "Cash", "Paypal", "Other"]
<br>
*method_freqs* = [270, 77, 32, 11]

```python
import matplotlib.pyplot as plt
method_type = ["Debit Card", "Cash", "Paypal", "Other"] 
method_freqs = [270, 77, 32, 11]
plt.pie(method_freqs)
plt.show()
```

The parts of the whole are calculated automatically and shown as a slice of the pie chart.  This is a nice looking pie chart, but let's use **plt.axis('equal')** to make it more round and less oblong.

```python
method_type = ["Debit Card", "Cash", "Paypal", "Other"] 
method_freqs = [270, 77, 32, 11]
plt.pie(method_freqs)
plt.axis('equal')
plt.show()
```

Great.  Now let's create labels for each slice of the pie chart by passing a *labels* argument within our **plt.pie( )** call.

```python
method_type = ["Debit Card", "Cash", "Paypal", "Other"] 
method_freqs = [270, 77, 32, 11]
plt.pie(method_freqs, labels = method_type)
plt.axis('equal')
plt.show()
```

The code was modified by setting the labels equal to the list of method types, *method_type*.  In doing this, we created labels for each slice.  Alternatively, we could have created the labels using the frequencies list, *method_freqs*.

```python
method_type = ["Debit Card", "Cash", "Paypal", "Other"] 
method_freqs = [270, 77, 32, 11]
plt.pie(method_freqs, labels = method_freqs)
plt.axis('equal')
plt.show()
```

Doing it this way, we should include a legend in order to distinguish what slice is what payment type.  In fact, let's add the features for a title and legend.

```python
method_type = ["Debit Card", "Cash", "Paypal", "Other"] 
method_freqs = [270, 77, 32, 11]
plt.pie(method_freqs, labels = method_freqs)
plt.axis('equal')
plt.title("Payment Method Frequency for Each Payment Type")
plt.legend(method_type)
plt.show()
```

Notice here, we passed into **plt.legend( )** the *method_type* list instead of explicitly typing out a list of legend text.
